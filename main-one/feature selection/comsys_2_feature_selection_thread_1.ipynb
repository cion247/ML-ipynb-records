# Filter method

```python
data = train.copy()
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
y = y.map({'Benign':0,'Malignant':1})
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from scipy.stats import pearsonr
```

```python

def select_top_features(x, y, num_features):
    if isinstance(x, pd.DataFrame):
        feature_labels = x.columns
    else:
        feature_labels = [f"Feature_{i}" for i in range(x.shape[1])]

    feature_correlations = []

    for feature_column in range(x.shape[1]):
        correlation, _ = pearsonr(x.iloc[:, feature_column], y)
        feature_correlations.append(correlation)

    # Sort feature labels based on absolute correlation values
    sorted_feature_labels = [label for _, label in sorted(zip(feature_correlations, feature_labels), key=lambda x: abs(x[0]), reverse=True)]

    # Select the top 'num_features' feature labels
    top_feature_labels = sorted_feature_labels[:num_features]

    return top_feature_labels


top_features = select_top_features(x, y, num_features=50)
top_features
```
